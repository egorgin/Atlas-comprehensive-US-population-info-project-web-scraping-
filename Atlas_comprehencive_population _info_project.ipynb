## State populations

```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import pandas as pd
import re
```

```python
my_url='https://statisticalatlas.com/United-States/Population#figure/state'
Uread=uReq(my_url)
map_html= Uread.read()
Uread.close()
map_soup=soup(map_html,'html.parser')
containers_1=map_soup.findAll('div',{'class':"col-sm-6 figure-col no-gutter"})
len(containers_1)
```

```python
labels_1= containers_1[4].findAll('text',{'font-family':"sans-serif"})
States_1=[]
for i in labels_1:
    i=str(i)
    States_1.append(re.split(">|<",i)[2])
#States_1    
#state_of_USA=[]    
#for i in state:
 #   states=i.replace(" ","-")
  #  state_of_USA.append(states)
#state_of_USA  
```

```python
percentage_range=containers_1[4].findAll('text',{'fill-opacity': "0.400"})
percentage_range
Population=[]
for i in percentage_range:
    i=str(i)
    Population.append(re.split(">|<",i)[2])
#Population
```

```python
statepop= pd.DataFrame({'States_1': States_1, 'Population':Population})
```
